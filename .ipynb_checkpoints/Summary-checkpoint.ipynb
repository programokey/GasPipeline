```python
import matplotlib.pyplot as plt
import pickle
```

```python
with open('gbdt_result.pickle', 'rb') as f:
    gbdt_result = pickle.load(f)
def to_latex(num):
    num = str(num)
    if 'e+' in num:
        base, exponent = num.split('e+')
        exponent = exponent.lstrip('0')
        num = '$%s\\times 10^{%s}$'%(base, exponent)
    elif 'e-' in num:
        base, exponent = num.split('e-')
        exponent = exponent.lstrip('0')
        num = '$%s\\times 10^{-%s}$'%(base, exponent)
    return num

def create_table(table):
    m, n = len(table), len(table[0])
    s = '''
\\begin{table}[hptb]
    \centering	
    \\begin{tabular}{|
    '''
    s += 'l|'*n + '} \hline'
    for row in table:
        s += '''
        '''
        s += '&'.join(map(lambda x: to_latex(x), row)) + '\\\\ \\hline'
    s += '''
        \end{tabular}
    \caption{{\small{yrdhfthftjnfy ghtfhgtmjh}}}
\end{table}
    '''
    return s
```

```python
num_leaves = gbdt_result['num_leaves']
num_leaves_number = [10, 100, 1000, 10000, 100000]
mse = [item['MSE'] for item in num_leaves]
mre = [item['MRE'] for item in num_leaves]
R_2 = [1 - item['R^2'] for item in num_leaves]

table = [['\\textbf{num\_leaves}'] + num_leaves_number, 
         ['\\textbf{MSE}'] + ['%.2g'%v for v in mse],
         ['\\textbf{MRE}'] + ['%g'%v for v in mre],
         ['\\textbf{$R^2$}'] + ['1 - %.4g'%v for v in R_2]]
print(mse)
print(R_2)
print(mre)
print(create_table(table))
```

```python
plt.scatter(x=num_leaves_number, y=mse, c='r') 
plt.plot(num_leaves_number, mse, scalex='log')
for x, y in zip(num_leaves_number, mse):
    plt.text(x, y + 10, '%.1f'%y, ha = 'center',va = 'bottom',fontsize=13)
plt.xscale('log')
plt.xlabel('num_leaves', fontsize=17)
plt.ylabel('MSE', fontsize=17)
plt.savefig('fig/gbdt_num_leaves_MSE.png',bbox_inches='tight')
plt.show()
```

```python
def draw_line_chart(x, y, x_label, y_label, format='%.g'):
    plt.scatter(x=x, y=y, c='r') 
    plt.plot(x, y, scalex='log')
    for x, y in zip(x, y):
        plt.text(x, y + 10, format%y, ha = 'center',va = 'bottom',fontsize=13)
    plt.xscale('log')
    plt.xlabel(x_label, fontsize=17)
    plt.ylabel(y_label, fontsize=17)
    plt.show()
```
